```python
#Question 3

Type 1 Error (False Positive):

Definition: Type 1 error occurs when we reject the null hypothesis when it is actually true. In other words, we conclude that there is an effect or a difference when there isn't one.



Example Scenario: A pharmaceutical company is testing a new drug for its effectiveness in treating a particular disease. The null hypothesis (H0) is that the drug has no effect on the disease. However, due to random chance or other factors, the study incorrectly concludes that the drug is effective (rejects H0) when, in fact, it doesn't work any better than a placebo.



Type 2 Error (False Negative):

Definition: Type 2 error occurs when we fail to reject the null hypothesis when it is actually false. In other words, we conclude that there is no effect or no difference when there actually is one.




Example Scenario: Continuing with the pharmaceutical example, let's say the drug being tested is actually effective in treating the disease. However, due to limitations in the study or other factors, the results fail to show a significant difference between the drug and the placebo. As a result, the study incorrectly fails to reject the null hypothesis (fails to reject H0), concluding that the drug is not effective when it actually is.
```
